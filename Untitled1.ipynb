```python
df = pd.read_csv("/content/dataset.csv", header=None, names=["head", "relation", "tail"], low_memory=False)

print("✅ Columns:", df.columns.tolist())
print("Rows:", len(df))
df.head()
```

```python
import networkx as nx


G = nx.DiGraph()

# Add edges from the triples
for _, row in df.iterrows():
    G.add_node(row['head'], type="entity")
    G.add_node(row['tail'], type="entity")
    G.add_edge(row['head'], row['tail'], relation=row['relation'])

print("✅ Knowledge Graph created")
print("Nodes:", G.number_of_nodes())
print("Edges:", G.number_of_edges())
```

```python
!pip install pyvis
from pyvis.network import Network

net = Network(height="750px", width="100%", notebook=True, bgcolor="#ffffff", font_color="black")

# Limit to a small sample (for speed)
sample_edges = list(G.edges(data=True))[:200]

for u, v, d in sample_edges:
    net.add_node(str(u), label=str(u))  # ✅ convert to string
    net.add_node(str(v), label=str(v))  # ✅ convert to string
    net.add_edge(str(u), str(v), label=str(d["relation"]))  # ✅ convert relation too

net.show("primekg_graph.html")
```

```python
def query_relation(entity_name, relation_type):
    results = []
    for u, v, data in G.edges(data=True):
        if u == entity_name and data['relation'] == relation_type:
            results.append(v)
    return results

# Example: find what a drug treats
drug = "DrugBank:DB00316"
print(f"\nEntities that {drug} treats:")
print(query_relation(drug, "treats"))

# Example: find what diseases a gene is associated with
gene = "Gene:TP53"
print(f"\nDiseases associated with {gene}:")
print(query_relation(gene, "associated_with"))
```
